## Menentukan Pertanyaan Bisnis

**SMART Question** adalah sebuah framework untuk merumuskan pertanyaan secara terstruktur agar memeperoleh informasi yang mendalam.

* **Spesific (Spesifik)**
  * Pertanyaan harus jelas, fokus pada sebuah topik tertentu, dan tidak bermakna ganda. Hindari pertanyaan yang terlalu luas.
    * Salah: Bagaimana cara meningkatkan penjualan?
    * Benar: Faktor apa saja yang memengaruhi penurunan penjualan produk kategori elektronik di wilayah Jakarta selama kuartal terakhir?
* **Measurable (Terukur)**
  * Pertanyaan harus bisa dijawab dengan angka atau matrix yang konkret, harus tahu apa yang akan dihitung.
    * Salah: Apakah pelanggan senang dengan layanan kita?
    * Benar: Berapa skor rata-rata Customer Satisfaction untuk layanan purna jual bulan ini dibandingkan bulan lalu?
* **Action-Oriented (Berorientasi Aksi)**
  * Hasil dari pertanyaan harus bisa memberikan arahan untuk melakukan tindakan nyata. Jika pertanyaan terjawab, stakeholder harus tahu apa langkah selanjutnya.
    * Salah: Mengapa orang suka berbelanja?
    * Benar: Fitur apa pada aplikasi yang paling sering digunakan sebelum pengguna memutuskan untuk melakukan checkout?
* **Relevant (Relevan)**
  * Hasil dari pertanyaan harus sejalan dengan tujuan utama bisnis atau masalah yang sedang dihadapi.
    * Salah: Menanyakan tentang stok gudang saat masalah utamanya adalah efektivitas kampanye media sosial.
    * Benar: Apakah kampanye iklan di Instagram memberikan Return on Ad Spend (ROAS) yang lebih tinggi dibandingkan iklan di TikTok?
* **Time-bound (Terikat Waktu)**
  * Pertanyaan harus ada batasan waktu yang jelas agar analisis memiliki konteks yang tepat.
    * Salah: Berapa banyak pengguna baru kita?
    * Benar: Berapa tingkat pertumbuhan pengguna baru secara bulanan (Month-over-Month) sepanjang tahun 2025?

**Contoh pertanyaan bisnis yang memenuhi seluruh elemen SMART**

***"Faktor apa saja yang memengaruhi penurunan conversion rate pada pengguna aplikasi Android di wilayah Jabodetabek sebesar 5% selama periode Flash Sale Maret 2026?"***

Keterangan:

- **Specific**: Fokus pada "penurunan conversion rate" untuk "aplikasi Android" di "Jabodetabek". Bukan sekadar penjualan turun.
- **Measurable**: Ada angka konkret yang ingin dianalisis, yaitu penurunan sebesar "5%".
- **Action-Oriented**: Dengan mengetahui faktor penyebabnya misalnya bug pada versi Android tertentu atau kendala logistik di Jabodetabek, tim bisa langsung melakukan perbaikan teknis atau operasional.
- **Relevant**: Penurunan konversi saat Flash Sale adalah masalah kritis bagi bisnis retail/e-commerce.
- **Time-bound**: Dibatasi pada periode spesifik "Flash Sale Maret 2026".

### Assessing Data

####  Identifying Missing Value Problem

```python
missing_count = df_raw.isnull().sum()
missing_pct   = (missing_count / len(df_raw) * 100).round(2)
missing_df = pd.DataFrame({
    'Missing Count': missing_count,
    'Missing (%)':   missing_pct
})
missing_df = missing_df[missing_df['Missing Count'] > 0].sort_values('Missing (%)', ascending=False)
print("=== Kolom dengan Missing Value ===")
print(missing_df.to_string())
```

**Steps to Take:**
- Kolom polutan (`PM2.5`, `PM10`, `SO2`, `NO2`, `CO`, `O3`) dan cuaca (`TEMP`, `PRES`, `DEWP`, `WSPM`) memiliki missing value.
- Karena ini data time-series per jam, missing value akan diisi menggunakan **interpolasi linear** per stasiun.
- Kolom `wd` (wind direction, bertipe string) akan diisi dengan **modus** per stasiun.

#### Identifying Duplicate Data Problem

```python
dup_cols = ['year', 'month', 'day', 'hour', 'station']
n_duplicates = df_raw.duplicated(subset=dup_cols).sum()
print(f"Jumlah baris duplikat (station + timestamp sama) : {n_duplicates}")
```

**Steps to Take:**
- Jika ada duplikat, hapus menggunakan `drop_duplicates()` dengan subset kolom identifikasi unik.

#### Identifying Invalid Value Problem (Nilai Negatif pada Polutan)

```python
pollutant_cols = ['PM2.5', 'PM10', 'SO2', 'NO2', 'CO', 'O3']

print("=== Cek Nilai Negatif pada Kolom Polutan ===")
found = False
for col in pollutant_cols:
    count = (df_raw[col] < 0).sum()
    if count > 0:
        print(f"  {col}: {count} nilai negatif")
        found = True
if not found:
    print("  Tidak ditemukan nilai negatif pada kolom polutan.")
```

**Insight:**
- Nilai negatif pada polutan tidak valid secara fisik — jika ditemukan akan diganti `NaN` lalu diimputasi.

#### Statistik Deskriptif

```python
df_raw[pollutant_cols + ['TEMP','PRES','DEWP','RAIN','WSPM']].describe().round(2)
```

### Cleaning Data

#### Fixing Duplicate Problem

```python
df_clean = df_raw.drop_duplicates(subset=['year', 'month', 'day', 'hour', 'station'])
print(f"Baris sebelum hapus duplikat : {len(df_raw):,}")
print(f"Baris sesudah hapus duplikat : {len(df_clean):,}")
print(f"Baris dihapus                : {len(df_raw) - len(df_clean):,}")
```

#### Fixing Invalid Value Problem (Nilai Negatif → NaN)

```python
for col in pollutant_cols:
    n_neg = (df_clean[col] < 0).sum()
    if n_neg > 0:
        df_clean.loc[df_clean[col] < 0, col] = np.nan
        print(f"  {col}: {n_neg} nilai negatif → NaN")

print(" Pengecekan nilai negatif selesai.")
```

#### Fixing Missing Value Problem (Interpolasi + Imputasi)

```python
df_clean = df_clean.copy()
df_clean['datetime'] = pd.to_datetime(df_clean[['year', 'month', 'day', 'hour']])
df_clean = df_clean.sort_values(['station', 'datetime']).reset_index(drop=True)

# Interpolasi linear per stasiun untuk kolom numerik
num_cols = pollutant_cols + ['TEMP', 'PRES', 'DEWP', 'RAIN', 'WSPM']
df_clean[num_cols] = (
    df_clean.groupby('station')[num_cols]
    .transform(lambda x: x.interpolate(method='linear', limit_direction='both'))
)

# Fallback: isi sisa NaN dengan median global
for col in num_cols:
    df_clean[col].fillna(df_clean[col].median(), inplace=True)

# Isi wind direction dengan modus per stasiun
df_clean['wd'] = df_clean.groupby('station')['wd'].transform(
    lambda x: x.fillna(x.mode()[0] if not x.mode().empty else 'N')
)

missing_after = df_clean.isnull().sum().sum()
print(f"Total missing value setelah cleaning : {missing_after}")
print(f"Shape akhir DataFrame bersih         : {df_clean.shape}")
```

#### Feature Engineering: Tambah Kolom Pendukung Analisis

```python
def get_season(month):
    if month in [3, 4, 5]:    return 'Spring'
    elif month in [6, 7, 8]:  return 'Summer'
    elif month in [9, 10, 11]: return 'Autumn'
    else:                      return 'Winter'

df_clean['season'] = df_clean['month'].map(get_season)
df_clean['season'] = pd.Categorical(
    df_clean['season'],
    categories=['Spring', 'Summer', 'Autumn', 'Winter'],
    ordered=True
)

# Kolom kategori AQI berdasarkan PM2.5 (standar EPA USA)
def get_aqi_category(pm25):
    if pd.isna(pm25):    return 'Unknown'
    elif pm25 <= 12.0:   return '1_Good'
    elif pm25 <= 35.4:   return '2_Moderate'
    elif pm25 <= 55.4:   return '3_Unhealthy for Sensitive'
    elif pm25 <= 150.4:  return '4_Unhealthy'
    elif pm25 <= 250.4:  return '5_Very Unhealthy'
    else:                return '6_Hazardous'

df_clean['AQI_Category'] = df_clean['PM2.5'].apply(get_aqi_category)

print(" Feature engineering selesai.")
print(f"Distribusi musim:")
print(df_clean['season'].value_counts().sort_index())
print(f"Distribusi AQI Category:")
print(df_clean['AQI_Category'].value_counts().sort_index())
```

**Insight:**
- Setelah cleaning, **0 missing value** tersisa di seluruh kolom.
- Kolom `season` dan `AQI_Category` ditambahkan sebagai fitur baru untuk keperluan analisis lanjutan.
- Kategori AQI mengikuti standar EPA USA yang umum digunakan dalam penelitian kualitas udara.

## Visualization & Explanatory Analysis

### Pertanyaan 1: Tren PM2.5 per Musim dan per Stasiun (2013–2017)

```python
# ── Persiapan data
season_station = (
    df_clean.groupby(['station', 'season'], observed=True)['PM2.5']
    .mean().reset_index()
    .rename(columns={'PM2.5': 'avg_PM25'})
)

station_order = (
    df_clean.groupby('station')['PM2.5'].mean()
    .sort_values(ascending=False).index.tolist()
)

palette  = {'Spring': '#4CAF50', 'Summer': '#FF9800', 'Autumn': '#795548', 'Winter': '#2196F3'}
seasons  = ['Spring', 'Summer', 'Autumn', 'Winter']

# ── Gambar
fig, axes = plt.subplots(2, 1, figsize=(14, 12))

# --- Plot atas: grouped bar per stasiun & musim ---
pivot = season_station.pivot(index='station', columns='season', values='avg_PM25').loc[station_order]
x     = np.arange(len(station_order))
width = 0.2

ax = axes[0]
for i, s in enumerate(seasons):
    ax.bar(x + i*width, pivot[s], width, label=s, color=palette[s], alpha=0.9, edgecolor='white')

ax.axhline(15, color='red', linestyle='--', linewidth=1.5, label='WHO Guideline (15 µg/m³)')
ax.set_xticks(x + width*1.5)
ax.set_xticklabels(station_order, rotation=35, ha='right', fontsize=9)
ax.set_ylabel('Rata-rata PM2.5 (µg/m³)', fontsize=11)
ax.set_title('Rata-rata Konsentrasi PM2.5 per Stasiun dan Musim Beijing, Maret 2013 – Februari 2017',
             fontsize=13, fontweight='bold', pad=12)
ax.legend(loc='upper right', framealpha=0.8, fontsize=9)
ax.yaxis.set_major_formatter(ticker.FormatStrFormatter('%.0f'))

# Anotasi stasiun terburuk
worst = station_order[0]
worst_max = pivot.loc[worst].max()
ax.annotate(f'Tertinggi: {worst}',
            xy=(x[0] + width*1.5, worst_max),
            xytext=(x[0] + width*1.5 + 1.5, worst_max + 6),
            arrowprops=dict(arrowstyle='->', color='#c0392b'),
            fontsize=8.5, color='#c0392b', fontweight='bold')

# --- Plot bawah: bar horizontal rata-rata tahunan ---
annual_avg = df_clean.groupby('station')['PM2.5'].mean().sort_values(ascending=True)
bar_colors = ['#e74c3c' if v > 75 else '#e67e22' if v > 55 else '#f1c40f' if v > 35 else '#2ecc71'
              for v in annual_avg.values]

ax2 = axes[1]
bars = ax2.barh(annual_avg.index, annual_avg.values, color=bar_colors, edgecolor='white', height=0.6)
ax2.axvline(15,  color='red',    linestyle='--', linewidth=1.5, label='WHO 15 µg/m³')
ax2.axvline(35,  color='orange', linestyle='--', linewidth=1.2, label='Moderate (35 µg/m³)')
ax2.axvline(75,  color='brown',  linestyle='--', linewidth=1.2, label='Unhealthy (75 µg/m³)')

for bar, val in zip(bars, annual_avg.values):
    ax2.text(val + 0.5, bar.get_y() + bar.get_height()/2,
             f'{val:.1f}', va='center', fontsize=8.5)

ax2.set_xlabel('Rata-rata PM2.5 (µg/m³)', fontsize=11)
ax2.set_title('Rata-rata PM2.5 Tahunan per Stasiun (Urutan Terburuk → Terbaik)',
              fontsize=12, fontweight='bold')
ax2.legend(fontsize=9)

plt.tight_layout(pad=3)
plt.savefig('viz_pertanyaan1.png', bbox_inches='tight', dpi=130)
plt.show()
```

## Analisis Lanjutan: Clustering Manual Berdasarkan Kategori AQI

Analisis lanjutan ini menerapkan teknik **Manual Clustering (Binning)** — mengelompokkan setiap jam pengukuran ke dalam 6 kategori kualitas udara berdasarkan standar **EPA USA**:

| Kategori | Rentang PM2.5 (µg/m³) | Makna |
|---|---|---|
| Good | 0 – 12.0 | Kualitas udara baik |
| Moderate | 12.1 – 35.4 | Kelompok sensitif perlu waspada |
| Unhealthy for Sensitive | 35.5 – 55.4 | Berbahaya bagi kelompok sensitif |
| Unhealthy | 55.5 – 150.4 | Berbahaya bagi semua orang |
| Very Unhealthy | 150.5 – 250.4 | Kondisi darurat kesehatan |
| Hazardous | > 250.4 | Kondisi berbahaya ekstrem |

**Tujuan:** Mengetahui berapa persen waktu setiap stasiun berada dalam kondisi berbahaya, dan bagaimana distribusinya berubah antar musim.


```python
# ── Hitung distribusi AQI per stasiun
aqi_order  = ['1_Good','2_Moderate','3_Unhealthy for Sensitive',
               '4_Unhealthy','5_Very Unhealthy','6_Hazardous']
aqi_labels = ['Good','Moderate','Unhealthy (Sensitive)','Unhealthy','Very Unhealthy','Hazardous']
aqi_colors = ['#2ecc71','#f1c40f','#e67e22','#e74c3c','#9b59b6','#8e44ad']

aqi_pct = (
    df_clean.groupby(['station','AQI_Category'])
    .size().unstack(fill_value=0)
    .reindex(columns=aqi_order, fill_value=0)
)
aqi_pct_norm = aqi_pct.div(aqi_pct.sum(axis=1), axis=0) * 100
aqi_pct_norm.columns = aqi_labels

# Urutkan berdasarkan % kondisi berbahaya
aqi_pct_norm['danger_score'] = aqi_pct_norm['Very Unhealthy'] + aqi_pct_norm['Hazardous']
aqi_pct_norm = aqi_pct_norm.sort_values('danger_score', ascending=False).drop('danger_score', axis=1)

print("=== Distribusi AQI Category per Stasiun (%) ===")
print(aqi_pct_norm.round(1).to_string())
```

```python
# ── Distribusi AQI per musim
aqi_season = (
    df_clean.groupby(['season','AQI_Category'], observed=True)
    .size().unstack(fill_value=0)
    .reindex(columns=aqi_order, fill_value=0)
)
aqi_season_norm = aqi_season.div(aqi_season.sum(axis=1), axis=0) * 100

print("=== Distribusi AQI Category per Musim (%) ===")
print(aqi_season_norm.round(1).to_string())
```

```python
import matplotlib.pyplot as plt
from matplotlib.patches import Patch

# ── Visualisasi Clustering AQI
fig = plt.figure(figsize=(16, 10))
gs  = fig.add_gridspec(2, 4, hspace=0.5, wspace=0.4)

# Stacked bar per stasiun
ax_bar = fig.add_subplot(gs[:, :2])
aqi_pct_norm.plot(kind='barh', stacked=True, color=aqi_colors,
                  ax=ax_bar, edgecolor='white', linewidth=0.5)
ax_bar.set_xlabel('Persentase Waktu (%)', fontsize=11)
ax_bar.set_title('Distribusi Kategori Kualitas Udara (AQI) per Stasiun Pemantauan',
                 fontsize=12, fontweight='bold')
ax_bar.legend(loc='lower right', fontsize=8.5, framealpha=0.8)
ax_bar.set_xlim(0, 105)
for i, (idx, row) in enumerate(aqi_pct_norm.iterrows()):
    danger = row['Very Unhealthy'] + row['Hazardous']
    ax_bar.text(101, i, f'{danger:.1f}%', va='center', fontsize=7.5, color='#8e44ad', fontweight='bold')
ax_bar.text(101, len(aqi_pct_norm), '% Bahaya', va='center', fontsize=7.5,
            color='#8e44ad', fontweight='bold', style='italic')

# Pie chart per musim
seasons_list = ['Spring','Summer','Autumn','Winter']
pie_positions = [(0,2),(0,3),(1,2),(1,3)]
for (row_idx, col_idx), season in zip(pie_positions, seasons_list):
    ax_pie = fig.add_subplot(gs[row_idx, col_idx])
    vals = aqi_season_norm.loc[season].values
    wedges, _, autotexts = ax_pie.pie(
        vals, colors=aqi_colors, autopct='%1.0f%%',
        startangle=90, pctdistance=0.75,
        textprops={'fontsize': 7}
    )
    for at in autotexts:
        if float(at.get_text().replace('%','')) < 4:
            at.set_text('')
    ax_pie.set_title(season, fontsize=11, fontweight='bold')

# Legend bersama untuk pie
legend_patches = [Patch(color=c, label=l.replace('',' '))
                  for c, l in zip(aqi_colors, aqi_labels)]
fig.legend(handles=legend_patches, loc='lower center', ncol=6,
           fontsize=8.5, bbox_to_anchor=(0.73, -0.02), framealpha=0.9)

plt.suptitle('Clustering AQI: Distribusi Kualitas Udara per Stasiun & Musim Beijing, 2013–2017',
             fontsize=13, fontweight='bold', y=1.02)
plt.savefig('viz_analisis_lanjutan.png', bbox_inches='tight', dpi=130)
plt.show()
```

```python
# ── Ringkasan persentase kondisi berbahaya
print("=== % Waktu dalam kondisi BERBAHAYA (Very Unhealthy + Hazardous) per Stasiun ===")
danger_pct = (aqi_pct_norm['Very Unhealthy'] + aqi_pct_norm['Hazardous']).sort_values(ascending=False)
for station, pct in danger_pct.items():
    bar = '█' * int(pct / 2)
    print(f"  {station:<25} {pct:5.1f}%  {bar}")

print()
print("=== % Waktu dalam kondisi AMAN (Good) per Musim ===")
good_season = aqi_season_norm[['Good']] if 'Good' in aqi_season_norm.columns else aqi_season_norm[['1_Good']]
print(good_season.round(1).to_string())
```

**Insight Analisis Lanjutan (Clustering AQI):**
- Rata-rata, stasiun di Beijing hanya memiliki kondisi udara "Good" kurang dari **5% waktu** — sangat mengkhawatirkan.
- Pada musim **Winter**, lebih dari **30% waktu** berada dalam kondisi "Very Unhealthy" atau "Hazardous".
- Pada musim **Summer**, proporsi berbahaya turun signifikan, namun "Unhealthy" masih mendominasi.
- **Wanshouxigong** dan **Dongsi** memiliki persentase waktu berbahaya tertinggi dibanding stasiun lain.
